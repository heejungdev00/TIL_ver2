```python
import pandas as pd
import numpy as np

# 1.

# (1) 임의로 두 개의 시리즈 객체를 만든다. 모두 문자열 인덱스를 가져야 하며 두 시리즈에 
# 공통적으로 포함되지 않는 라벨이 있어야 한다.
df1 = pd.Series([9904312, 3448737, 2890451, 2466052],index=["서울", "부산", "Ohio", "Texas"])
df2 = pd.Series({"Ohio": 35000, "Texas": 71000, "Oregon": 16000, "Utah": 5000})

df1

df2

# (2) 위에서 만든 두 시리즈 객체를 이용하여 사칙 연산을 한다.

hap = df1 + df2

hap

cha = df1 - df2
cha

gop = df1 * df2
gop

na = df1 / df2
na

# 2.
# 다음 조건을 만족하는 임의의 데이터프레임을 하나 만든다.

# (1) 열의 갯수와 행의 갯수가 각각 5개 이상이어야 한다.

# (2) 열에는 정수, 문자열, 실수 자료형 데이터가 각각 1개 이상씩 포함되어 있어야 한다.

data={
    'a': [1,2,3,4,5],
    'b': [1.1,1.3,13.11,333.221,55.55422],
    'c': ['a','s','d','f','g'],
    'd': [3,4,5,6,7],
    'e': ['s','d','f','g','h'] 
}
df = pd.DataFrame(data)
df

# 3.
# 다음 데이터프레임에서 지정하는 데이터를 뽑아내거나 처리하라.

data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)

df

# (1) 모든 학생의 수학 점수를 시리즈로 나타낸다.
df.수학

# (2) 모든 학생의 국어와 영어 점수를 데이터 프레임으로 나타낸다.
df.loc[:,['국어', '영어']]

# (3) 모든 학생의 각 과목 평균 점수를 새로운 열로 추가한다.
df['평균점수']=df.mean(axis=1)
df

# (4) 방자의 영어 점수를 80점으로 수정하고 평균 점수도 다시 계산한다.
df.loc['방자','영어']=80

df.loc['방자','평균점수']=df.loc['방자'][:3].mean()

df

# (5) 춘향의 점수를 데이터프레임으로 나타낸다.

df.loc[['춘향']]

# (6) 향단의 점수를 시리즈로 나타낸다.

df.loc['향단']
```
