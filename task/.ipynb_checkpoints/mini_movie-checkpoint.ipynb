```python
response = requests.get(url)
xml_data=response.content
soup=BeautifulSoup(xml_data, 'xml')
```

```python
movies = soup.find_all('movie')
```

```python
movies_list=[]
```

```python
movies_list = []

for movie in movies:
    movieCd_tag = movie.find('movieCd') #영화코드
    movieCd = movieCd_tag.text if movieCd_tag is not None else None

    movieNm_tag = movie.find('movieNm') # 영화명(국문)
    movieNm = movieNm_tag.text if movieNm_tag is not None else None

    movieNmEn_tag = movie.find('movieNmEn') # 영화명(영문)
    movieNmEn = movieNmEn_tag.text if movieNmEn_tag is not None else None

    openDt_tag = movie.find('openDt') # 개봉일
    openDt = openDt_tag.text if openDt_tag is not None else None

    nationAlt_tag = movie.find('nationAlt')# 제작국가
    nationAlt = nationAlt_tag.text if nationAlt_tag is not None else None

    genreAlt_tag = movie.find('genreAlt')# 장르
    genreAlt = genreAlt_tag.text if genreAlt_tag is not None else None

    peopleNm_tag = movie.find('peopleNm')# 감독
    peopleNm = peopleNm_tag.text if peopleNm_tag is not None else None

    movies_list.append({'movieCd': movieCd, 'movieNm': movieNm, 'movieNmEn': movieNmEn,
                        'openDt': openDt, 'nationAlt': nationAlt, 'genreAlt': genreAlt, 'peopleNm': peopleNm})
```

```python
# dataframe 입력
import pandas as pd
db_result_df=pd.DataFrame(db_export)
db_result_df
```

```python
# 고유한 장르 목록을 확인하는 코드
unique_genres = db_result_df['장르'].unique()
genreList = []
for genre in unique_genres:
    try:
        x = genre.split(',')
        genreList.extend(x)
    except AttributeError:
        pass

genreList=list(set(genreList)) # 장르의 종류
genreList.remove('기타')
```

```python
genreList
```

```python
import random

# 장르별로 하나의 영화를 추천하는 함수
def recommend_movie_by_genre(genreList, df):
    recommended_movies = []
    for genre in genreList:
        # 해당 장르에 속하는 영화들 필터링
        genre_movies = df[df['장르'].str.contains(genre, na=False)]
        if not genre_movies.empty:
            # 장르에 속하는 영화 중에서 랜덤으로 하나 선택
            random_movie = genre_movies.sample(1)
            recommended_movies.append(random_movie)
    return recommended_movies

# 장르별로 영화 추천
recommended_movies = recommend_movie_by_genre(genreList, db_result_df)

# 결과 출력
print("장르별 추천 영화:")
for movie in recommended_movies:
    print(movie[['영화명', '장르']])
```

```python
import random

# 각 제작연도별로 하나의 영화를 추천하는 함수
def recommend_movie_by_year(df):
    # 데이터프레임을 제작연도로 그룹화
    grouped_by_year = df.groupby('제작연도')
    
    # 추천할 영화 리스트
    recommended_movies = []
    
    # 각 그룹에서 랜덤하게 하나의 영화를 선택하여 리스트에 추가
    for year, group in grouped_by_year:
        movie = group.sample(n=1)
        recommended_movies.append(movie)
    
    return recommended_movies

# 추천 영화 출력
recommended_movies = recommend_movie_by_year(df)
for movie in recommended_movies:
    print(movie)
```
