```python
# 1. 동전을 20번 던져 앞면(숫자 1)과 뒷면(숫자 0)이 나오는 실험을 파이썬으로 작성하여 출력해보자(numpy)
coin = [np.random.randint(2) for i in range(20)]
coin

import numpy as np

# 2. 주사위를 10000번 던져서 나오는 숫자의 평균을 구하라.(numpy)

dice=[np.random.randint(1,7) for i in range(10000)]

np.mean(dice)

# 3. 난수로 이루어진 4 x 5 형태의 데이터 행렬을 만들고 이 데이터에 대해 다음과 같은 값을 구한다.(numpy)

data = np.random.rand(4, 5)

data

# 전체의 최댓값
data.max()

# 각 행의 합
np.sum(data, axis=1)

# 각 행의 최댓값
np.max(data, axis=1)

# 각 열의 평균
np.mean(data, axis=0)

# 각 열의 최솟값
np.min(data, axis=0)

# 4.  다음 데이터를 시리즈 구조로 정의하고 '광주' 도시의 인구를 참조하시오
# 서울-1000, 부산-300, 인천-200, 광주-100

import pandas as pd
data=pd.Series([1000,300,200,100], index=['서울','부산','인천','광주'])

data.index.name="도시"

data

data['광주']

# 5. 다음의 Series를 생성하고 sample 변수에 대입하고 출력하세요
# [출력]
# 가    10
# 나    20
# 다    30
# 라    40
# 마    50
# dtype: int64

sample=pd.Series([i for i in range(10,51,10)], index=['가','나','다','라','마'])

sample

# 6. sample중 '나'와 '라' 데이터를 조회하세요
sample[['나','라']]

np.random.seed(20)
sample2 = pd.Series(np.random.randint(100, 200, size=(15,)))
sample2

# 7. sample2 중 160 이하인 데이터만 필터하세요

sample2[sample2<=160]

# 8. sample2 중 130 이상 170 이하인 데이터만 필터하세요

sample2[(sample2>=130) & (sample2<=170)]

# 9.다음과 같은 Series를 생성해 주세요
# [출력]
# 가     apple
# 나       NaN
# 다    banana
# 라      kiwi
# 마    gubong
# dtype: object

res=pd.Series(['apple', np.nan, 'banana', 'kiwi', 'gubong'], index=['가','나','다','라','마'])

res

# 10. 
# - 숫자로 이루어진 리스트에서 사용자가 입력한 숫자를 검색하는 모듈을 다음 요건에 따라 만들어 보자
# - 검색 모듈은 선형 검색 알고리즘을 이용하자
# - 리스트는 1부터 20까지의 정수 중에서 난수 10개를 이용한다.
# -검색에 성공하면 해당 정수의 인덱스를 출력하고, 검색 결과가 없다면 -1을 출력하자

def func(num):
    li=[np.random.randint(1,21) for i in range(10)]
    print(li)
    for i,v in enumerate(li):
        if v==num:
            return i
    return -1

func(5)
```
