```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import re

driver=webdriver.Chrome()

url='https://www.starbucks.co.kr/store/store_map.do?disp=locale'

driver.get(url)

select_btn='#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'

driver.find_element('css selector', select_btn).click()

s_all_btn = '#mCSB_2_container > ul > li:nth-child(1) > a'

driver.find_element('css selector',s_all_btn).click()

html=driver.page_source

soup=BeautifulSoup(html,'html.parser')

starbucks_list=soup.select("#mCSB_3_container > ul > li")

starbucks_list_con=soup.select("li.quickResultLstCon")

starbucks_list[0]

starbucks_addr=[]
for item in starbucks_list:
    addr=str(item.select("p")[0]).split("<br/>")[0].split(">")[1]
    starbucks_addr.append(addr)
starbucks_addr

count=0
gu_lst=[]
for addr in starbucks_addr:  # 주소에서 구만 뽑음.
    gu=addr.split(' ')[1]
    gu_lst.append(gu)

gu_lst_set=sorted(list(set(gu_lst)))  # 구의 종류
print(gu_lst_set)

res=[]
for gu in gu_lst:
    for name in gu_lst_set:
        if name in gu:
            s=name+'-'+str(gu_lst.count(name))
            res.append(s)
res=sorted(list(set(res)))   # ex 강남구-91, 강동구-17.... 로 출력
print(res)

max=0
for r in res:
    if max < int(r.split('-')[1]):
        max=int(r.split('-')[1])
        max_gu=r.split('-')[0]

print('스타벅스 매장이 가장 많은 구는 {}이고 {}개가 있다.'.format(max_gu,max))
        

min=max
for r in res:
    if min > int(r.split('-')[1]):
        min=int(r.split('-')[1])
        min_gu=r.split('-')[0]

print('스타벅스 매장이 가장 적은 구는 {}이고 {}개가 있다.'.format(min_gu,min))

hap=0
for r in res:
    hap+=int(r.split('-')[1])
avg_starbucks=hap/len(gu_lst_set)
print('서울에 있는 구에는 평균 {}개의 스타벅스 매장이 있다.'.format(avg_starbucks))
```
