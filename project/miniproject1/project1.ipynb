```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time
import random

en='''owing
elect
ablaze
adaptable
indestructible
this
those
three
through
throughout
tough
toward
town
trade
traditional
training
travel
treat
treatment
tree
trial
trip
trouble
true
truth
try
turn
TV
two
type
under
understand
unit
until
up
upon
us
use
usually
value
various
very
victim
view
wait
walk
wall
want
war
watch
water
weapon
wear
week
weight
well
west
western
what
whatever
when
where
whether
which
social
society
soldier
some
somebody
son
song
sort
process
produce
product
production'''

en_lst=en.split('\n')
kr_lst=[]  # 7분정도 걸리는 코드.(기본 사전 생성)

driver=webdriver.Chrome()            
time.sleep(3)
url='https://papago.naver.com/'
for e in en_lst:
    driver.get(url)
    time.sleep(1)
    input_en=driver.find_element('css selector', '#txtSource')
    input_en.send_keys(e)
    time.sleep(3)
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    kr_lst.append(soup.select('#txtTarget > span')[0].text)

basic_dic=dict(zip(en_lst,kr_lst))

def get_trans(s):    # 영어단어 검색해서  뜻을 반환하는 함수
    driver.get(url)
    time.sleep(1)
    input_en=driver.find_element('css selector', '#txtSource')
    input_en.send_keys(s)
    time.sleep(1)
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    
    return soup.select('#txtTarget > span')[0].text

def read_basic_dic():   # 기본 사전 읽어주는 함수
    print(basic_dic)

quiz_on=True
def word_quiz():   # 기존 사전에 있는 영단어로 test
    global quiz_on
    score=0
    quiz_num=0
    while quiz_on:
        quiz_num += 1
        quiz_choice = random.sample(en_lst,4)
        answer_index = random.randint(0,3)
        print()
        print('{}의 뜻은?'.format(quiz_choice[answer_index]))
    
        for i in range(4):
            print(f'{i+1}. {basic_dic.get(quiz_choice[i])}')
        
        user_input = int(input('정답을 입력해주세요. 종료하려면 0을 입력. \n'))
    
        if user_input == 0:
            quiz_on=False
            print('종료되었습니다. 총 {}문제 중 {}를 맞췄습니다'.format(quiz_num-1,score))
            print()
        elif user_input == answer_index + 1:
            score+=1
            print('정답입니다.')
        else: 
            print('틀렸습니다. 정답은 {}번입니다'.format(answer_index + 1))

def get_en(s):   # 파파고에 영단어 검색색
    driver.get(url)
    time.sleep(1)
    input_kr=driver.find_element('css selector', '#txtSource')
    input_kr.send_keys(s)
    time.sleep(1)
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
    
    return soup.select('#txtTarget > span')[0].text.strip()

user=0
en_record=[]
kr_record=[]
while True:
    start_m='''영어사전프로그램입니다. 어떤 것을 원하시나요?\n
    1. 기본 영어사전보기
    2. 영어단어 test
    3. 영단어 검색하기
    4. 한영사전 검색하기
    0. 프로그램 종료하기'''
    print(start_m)
    user=int(input())
    if user==0:
        print('프로그램을 종료합니다.')
        break
    
    elif user==1:
        user_dic=int(input('기본영어사전입니다. 어떤 것을 원하시나요?\n 1.전체사전보기\n 2.사전에 단어검색하기\n 0.돌아가기'))
        if user_dic==1:      
            read_basic_dic()
        elif user_dic==2 :
            user_get_word=input('사전에서 검색할 단어를 입력하시오')
            print(basic_dic.get(user_get_word,'기본사전에 단어가 없습니다.'))
            print()
        else:
            print('돌아가기')
    
    elif user==2:
        quiz_on=True
        word_quiz()
    
    elif user==3:
        user_en_=int(input('1.단어검색하기\n2.검색기록확인하기\n0.돌아가기'))
        if user_en_==1:
            word_input=input('검색하고싶은 영어단어를 입력하세요.')
            en_record.append(word_input)
            driver=webdriver.Chrome()
            time.sleep(3)
            url='https://papago.naver.com/'
            print('{}의 뜻은 {}입니다.'.format(word_input,get_trans(word_input)))
            user_Y_N=int(input('기본 영어사전에 추가하시겠습니까?\n 1.추가하겠습니다\n 2.아니요 종료해주세요.\n'))
            if user_Y_N==1:
                basic_dic[word_input]=get_trans(word_input)
                print('추가되었습니다.')
            else:
                print('네 종료하겠습니다.')
        elif user_en_==2:
            print(en_record)
        else:
            print('돌아가기')
    elif user==4:
        user_kr_=int(input('1.단어검색하기\n2.검색기록확인하기\n0.돌아가기'))
        if user_kr_==1:
            kr_input=input('영어로 알고싶은 단어를 입력하세요.')
            kr_record.append(kr_input)
            driver=webdriver.Chrome()
            time.sleep(3)
            url='https://papago.naver.com/?sk=ko&tk=en'
            print('{}는 영어로 {}입니다.'.format(kr_input,get_trans(kr_input)))
        elif user_kr_==2:
            print(kr_record)
        else:
            print('돌아가기')
```
