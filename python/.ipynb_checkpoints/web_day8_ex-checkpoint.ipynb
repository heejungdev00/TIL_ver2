```python
# 1. 사용자 입력과 random함수를 사용하여, 사용자(User)와 컴퓨터가 대결하는 가위 바위 보 게임을 만들어보자.
# 입력: [문자열] "가위", "바위" 혹은 "보" 출력: [문자열] 결과 반환

user=input('가위 바위 보 중 입력: ')

game=['가위', '바위', '보']

import random
com=random.choice(game)
print('com은',com)

if user=='가위':
    if com=='가위':
        print('비김')
    elif com=='바위':
        print('사용자가 졌음')
    else:
        print('사용자가 이김')
        
elif user=='바위':
    if com=='가위':
        print('사용자가 이김')
    elif com=='바위':
        print('비김')
    else:
        print('사용자가 짐')
        
elif user=='보':
    if com=='가위':
        print('사용자가 졌음')
    elif com=='바위':
        print('사용자가 이김')
    else:
        print('비김')

# 2. 2^15 = 32768 의 각 자리수를 더하면 3 + 2 + 7 + 6 + 8 = 26 입니다. 2^1000의 각 자리수를 모두 더하면 얼마입니까?

ss=str(2**1000)
hap=0
for s in ss:
    hap+=int(s)

print(hap)

# 3. 10진수를 n진수로 변환하는 프로그램을 작성하시오.. (단, n의 범위는 2 <= n <= 16)
# 예)
# 2진수로 변환 : 23310 --> 111010012
# 8진수로 변환 : 23310 --> 3518
# 16진수로 변환 : 23310 --> E916

# num=23310
num=int(input())
n=int(input())

n_16_dict={'10':'A', '11':'B', '12':'C', '13':'D', '14':'E', '15':'F'}

if n==2 or n==8:
    res=[]
    while True:
        if num>=n:
            res.append(num%n)
            num=num//n
        else:
            res.append(num) 
            break
elif n==16:
    res=[]
    while True:
        if num>=n:
            if num%n>=10:
                res.append(n_16_dict.get(str(num%n)))
                num=num//n
            else:
                res.append(str(num%n))
                num=num//n
        else:
            if num>=10:
                res.append(n_16_dict.get(str(num)))
                break
            else:
                res.append(str(num))
                break

print(res)

if n==16:
    n_num=''.join(list(reversed(res)))
    print(n_num)
else:
    n_num=int(''.join(map(str,list(reversed(res)))))
    print(n_num)

# 4. 문자열 하나를입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하라.
# make_url("naver")
# www.naver.com

def make_url(s):
    res='www.'+s+'.com'
    return res

print(make_url('naver'))

# 5. 문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하라.
# make_list("abcd")
# ['a', 'b', 'c', 'd']

def make_list(ss):
    res=[s for s in ss]
    return res

print(make_list("abcd"))

# 6. 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.

# pickup_even([3, 4, 5, 6, 7, 8])
# [4, 6, 8]

def pickup_even(lst):
    res=[s for s in lst if s%2==0]
    return res

print(pickup_even([3, 4, 5, 6, 7, 8]))



# 7. 콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하라.
# convert_int("1,234,567")
# 1234567

def convert_int(s):
    return int(''.join(s.split(',')))

print(convert_int("1,234,567"))
```
